```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
df.columns
```

1. **Basic Data Exploration**

```python
# 1. Identify the number of rows and columns
num_rows, num_cols = df.shape
print(f"Number of rows: {num_rows}")
print(f"Number of columns: {num_cols}")

# 2. Determine the data types of each column
data_types = df.dtypes
print("\nData types of each column:")
print(data_types)

# 3. Check for missing values in each column
missing_values = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values)
```

2. **Descriptive Statistics**

```python
# 1. Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().values[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
salary_range = max_salary - min_salary
std_dev_salary = df['TotalPay'].std()

# 2. Print the results
print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

3. **Data Cleaning**

```python
# Before handling missing values
print("Missing values before handling:")
print(df.isnull().sum())

# Impute missing values with the mean of the 'TotalPay' column
mean_totalpay = df['TotalPay'].mean()
df['TotalPay'].fillna(mean_totalpay, inplace=True)

# After handling missing values
print("\nMissing values after handling:")
print(df.isnull().sum())
```

**Reasons for using mean imputation**:

Preservation of Mean: Imputing with the mean helps to preserve the overall mean of the dataset, which can be important in maintaining the statistical characteristics of the variable.

Simplicity: Mean imputation is straightforward and easy to implement, making it a quick solution for handling missing values.

4. **Basic Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# a histogram for the distribution of salaries (TotalPay)
plt.figure(figsize=(10, 6))
sns.histplot(df['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# a pie chart for the proportion of employees in different departments
plt.figure(figsize=(8, 8))
department_counts = df['JobTitle'].value_counts().head(10)  # Adjust as needed
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

5. **Grouped Analysis**

```python
# Group data by 'JobTitle' and calculate average salary for each group
average_salary_by_jobtitle = df.groupby('JobTitle')['TotalPay'].mean().sort_values(ascending=False)

# Print the results
print("Average Salary by Job Title:")
print(average_salary_by_jobtitle)
```

6. **Simple Correlation Analysis**

```python
# numerical column
numerical_column = 'BasePay'

# Check for correlation between 'TotalPay' and the selected numerical column
correlation = df['TotalPay'].corr(df[numerical_column])

# Print the correlation coefficient
print(f"Correlation between TotalPay and {numerical_column}: {correlation}")

# Plot a scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df[numerical_column], y=df['TotalPay'])
plt.title(f'Scatter Plot: TotalPay vs {numerical_column}')
plt.xlabel(numerical_column)
plt.ylabel('Total Pay')
plt.show()
```
